```python
import pandas as pd
import numpy as np
import sqlite3
import sqlalchemy
from sqlalchemy import create_engine
```

```python
file_path = "../../datasets/air_carriers.csv"
```

```python
df = pd.read_csv(file_path, names=['IATACode', 'ICAOCode', 'AirCarrierName'], header=0)
```

```python
air_carriers = df.set_index('IATACode')
```

```python
air_carriers
```

```python
db_path = "../../dbs/delays.db"
```

```python
conn = sqlite3.connect(db_path)
conn.execute("DROP TABLE IF EXISTS air_carriers;")

create_query = ''' \
            CREATE TABLE air_carriers
            ( 
                IATACode                       VARCHAR, 
                ICAOCode                       VARCHAR,
                AirCarrierName                 VARCHAR
            );
    '''
conn.execute(create_query)
conn.close()
engine = sqlalchemy.create_engine("sqlite:///"+db_path, echo=False)
```

```python
engine = sqlalchemy.create_engine("sqlite:///"+db_path, echo=False)
air_carriers.to_sql('air_carriers', con=engine, if_exists='append', index=False)
```
